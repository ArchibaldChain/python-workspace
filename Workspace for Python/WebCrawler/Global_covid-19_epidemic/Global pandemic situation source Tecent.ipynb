```python
data_dic = response.json()
data_dic
```

```python
import json

data_dict = json.loads(data_dic['data'])  # loads the dictionary from the value
```

```python
data_dict['foreignList']
```

```python
data_dict.keys()
```

```python
data_dict['globalDailyHistory']
```

```python
data_dict['foreignList'][12]
```

```python
country_list= []
children_list = []
for country in data_dict['foreignList']:
    if country.get('children') :  # we can't use country['children']
        children = country.pop('children', None)  # if not return None
        children_list.append(children)
    country_list.append(country)
        
```

```python
country_list
```

```python
import pandas as pd
cou_df = pd.DataFrame(country_list)
```

```python
cou_df
```
