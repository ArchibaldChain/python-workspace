```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
%matplotlib inline
```

```python
data = "./gravity_wave_file/"
rate_h, hstrain = wavfile.read(data+ r"H1_Strain.wav", "rb")
rate_l, lstrain = wavfile.read(data+"L1_Strain.wav", "rb")
```

```python
plt.plot(hstrain)
```

```python
plt.plot(lstrain)
```

```python
import pandas as pd
reft = pd.read_csv(data+'wf_template.txt',
                  names = ['refttime', 'reft_H1'], delimiter = ' ')
```

```python
reft
```

### get X axis

```python
htime_interval = 1/rate_h
ltime_interval = 1/rate_l
```

```python
htime_len = hstrain.shape[0]/rate_h
htime = np.arange(-htime_len/2, htime_len/2, htime_interval)
ltime_len = lstrain.shape[0]/rate_l
ltime = np.arange(-ltime_len/2, ltime_len/2, ltime_interval)
```

### Pating H1 Strain

```python
fig = plt.figure(figsize=(12,6))
plth = fig.add_subplot(2,2,1)
plth.plot(htime,hstrain, 'g')
plth.set_xlabel('Time (seconds)')
plth.set_ylabel('H1 Strain')
plth.set_title('H1 Strain')
```

```python
pltl = fig.add_subplot(2,2,2)
pltl.plot(ltime,lstrain, 'g')
pltl.set_xlabel('Time (seconds)')
pltl.set_ylabel('L1 Strain')
pltl.set_title('L1 Strain')
```

### Pating reft Strain

```python
pltref = fig.add_subplot(2,1,2)
pltref.plot(reft['refttime'],reft['reft_H1'], 'g')
pltref.set_xlabel('Time (seconds)')
pltref.set_ylabel('Template Strain')
pltref.set_title('Template')
```

```python
fig.tight_layout() # adjust the margin of the figure
```

```python
#plt.savefig("Gravitational_Waves_Original.png", dpi = 600)
plt.show()
plt.close(fig)
```

### Putting together

```python

fig = plt.figure(figsize=(12,6))
plth = fig.add_subplot(2,2,1)
plth.plot(htime,hstrain, 'g')
plth.set_xlabel('Time (seconds)')
plth.set_ylabel('H1 Strain')
plth.set_title('H1 Strain')


pltl = fig.add_subplot(2,2,2)
pltl.plot(ltime,lstrain, 'g')
pltl.set_xlabel('Time (seconds)')
pltl.set_ylabel('L1 Strain')
pltl.set_title('L1 Strain')

pltref = fig.add_subplot(2,1,2)
pltref.plot(reft['refttime'],reft['reft_H1'], 'g')
pltref.set_xlabel('Time (seconds)')
pltref.set_ylabel('Template Strain')
pltref.set_title('Template')

fig.tight_layout()

plt.savefig("Gravitational_Waves_Original.png", dpi = 600)
plt.show()
plt.close(fig)
```
